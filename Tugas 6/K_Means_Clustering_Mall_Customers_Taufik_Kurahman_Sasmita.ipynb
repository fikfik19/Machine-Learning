```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Path ke file CSV
file_path = "/content/drive/MyDrive/Dataset/Mall_Customers.csv"

# Membaca dataset
dataset = pd.read_csv(file_path)

# Pilih fitur yang akan digunakan untuk clustering
features = dataset[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Normalisasi data
features_normalized = (features - features.mean()) / features.std()

# Menentukan jumlah cluster (dalam contoh ini, kita akan menggunakan 5 cluster)
num_clusters = 5

# Melakukan clustering dengan algoritma K-Means
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
clusters = kmeans.fit_predict(features_normalized)

# Menambahkan kolom 'Cluster' ke dalam DataFrame
dataset['Cluster'] = clusters

# Menampilkan hasil clustering
print(dataset.head())

# Visualisasi hasil clustering
plt.scatter(dataset['Annual Income (k$)'], dataset['Spending Score (1-100)'], c=dataset['Cluster'], cmap='viridis')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Clustering Result')
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Path ke file CSV
file_path = "/content/drive/MyDrive/Dataset/Mall_Customers.csv"

# Membaca dataset
dataset = pd.read_csv(file_path)

# Pilih fitur yang akan digunakan untuk clustering
features = dataset[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Normalisasi data
features_normalized = (features - features.mean()) / features.std()

# Range jumlah klaster yang akan diuji
k_range = range(1, 11)

# List untuk menyimpan inertia untuk setiap jumlah klaster
inertia_values = []

# Melakukan clustering dengan berbagai jumlah klaster dan menghitung inertia
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features_normalized)
    inertia_values.append(kmeans.inertia_)

# Menampilkan plot inertia vs jumlah klaster
plt.plot(k_range, inertia_values, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Inertia vs Number of Clusters')
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Path ke file CSV
file_path = "/content/drive/MyDrive/Dataset/Mall_Customers.csv"

# Membaca dataset
dataset = pd.read_csv(file_path)

# Pilih fitur yang akan digunakan untuk clustering
features = dataset[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Normalisasi data
features_normalized = (features - features.mean()) / features.std()

# Jumlah klaster yang diinginkan
num_clusters = 5

# Melakukan clustering dengan algoritma K-Means
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
clusters = kmeans.fit_predict(features_normalized)

# Menambahkan kolom 'Cluster' ke dalam DataFrame
dataset['Cluster'] = clusters

# Visualisasi hasil clustering
plt.figure(figsize=(10, 6))

# Memvisualisasikan klaster menggunakan scatter plot
for cluster_label in range(num_clusters):
    cluster_data = dataset[dataset['Cluster'] == cluster_label]
    plt.scatter(cluster_data['Annual Income (k$)'], cluster_data['Spending Score (1-100)'], label=f'Cluster {cluster_label}')

plt.title('Clustering Result')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Path ke file CSV
file_path = "/content/drive/MyDrive/Dataset/Mall_Customers.csv"

# Membaca dataset
dataset = pd.read_csv(file_path)

# Pilih fitur yang akan digunakan untuk clustering
features = dataset[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Normalisasi data
features_normalized = (features - features.mean()) / features.std()

# Range jumlah klaster yang akan diuji
k_range = range(2, 11)

# List untuk menyimpan silhouette score untuk setiap jumlah klaster
silhouette_scores = []

# Melakukan clustering dengan berbagai jumlah klaster dan menghitung silhouette score
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    clusters = kmeans.fit_predict(features_normalized)
    silhouette_avg = silhouette_score(features_normalized, clusters)
    silhouette_scores.append(silhouette_avg)

# Menampilkan silhouette score untuk setiap jumlah klaster
for k, silhouette_score in zip(k_range, silhouette_scores):
    print(f"Silhouette Score for {k} clusters: {silhouette_score}")
```
