```python
# Membaca library
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
file_path = "/content/drive/MyDrive/Dataset/Mall_Customers.csv"
data = pd.read_csv(file_path)
```

```python
# Pilih fitur yang akan digunakan untuk clustering
X = data.iloc[:, [3, 4]].values  # Memilih kolom 'Annual Income (k$)' dan 'Spending Score (1-100)'
```

```python
# Hierarchical Clustering
model = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
clusters = model.fit_predict(X)
```

```python
# Visualisasi hasil clustering
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X[:,0], y=X[:,1], hue=clusters, palette='viridis')
plt.title('Hierarchical Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```
